##### **Problem 04:**
Generate the vector `[36, 25, 16, 9, 4, 1]` and then find the square root of each element of the vector using for loop. **Hint:** Use `np.sqrt()` function to find the square root of a number, don't use any other built-in function.

```python
vector = np.array([36, 25, 16, 9, 4, 1])


square_roots = np.empty(vector.shape, dtype=float)


for i in range(vector.size):
    square_roots[i] = np.sqrt(vector[i])

print("Original vector:", vector)
print("Square roots of each element:", square_roots)
```

##### **Problem 05:**
Write a function that takes a number and returns a vector like the previous problem. For example, if the input is 4, then the output should be `[16, 9, 4, 1]`. If the input is 10, then the output should be `[100, 81, 64, 49, 36, 25, 16, 9, 4, 1]`.

```python
def generate_square_vector(n):
    if n <= 0:
        return np.array([])  # Return an empty array for invalid input

    vector = np.arange(n, 0, -1) ** 2

    return vector

input_1 = 4
input_2 = 10

output_1 = generate_square_vector(input_1)
output_2 = generate_square_vector(input_2)

print(f"For input {input_1}, the output is: {output_1}")
print(f"For input {input_2}, the output is: {output_2}")





```

##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-5, 5]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_range = np.arange(-5, 5.5, 0.5)
y_range = np.arange(-5, 5.5, 0.5)


X, Y = np.meshgrid(x_range, y_range)


Z = 1 - X - Y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot_surface(X, Y, Z, cmap='viridis')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


ax.set_title('Surface Plot of x + y + z = 1')


plt.show()
```
